```python
webpage=requests.get('https://data.who.int/countries/')
```

```python
page=BeautifulSoup(webpage.text,'lxml')
```

```python
page
```

```python
page.find_all('div',class_='alphabetical-box')[0].find_all('a')[0].attrs['href']
```

```python
Link=[]
for i in page.find_all('div',class_='alphabetical-box'):
  for j in i.find_all('a'):
    # print(j.attrs['href'])
    Link.append(j.attrs['href'])
```

```python
def get_webpage(url):
  return requests.get(url).text
```

```python
len(Link)
```

```python
Name=[]
Population=[]
CHE=[]
WHO_REG=[]
WBIL=[]
counter=1
for l in Link:
  page1=BeautifulSoup(get_webpage(l),'lxml')
  Country_box=page1.find_all('div',class_='col-lg-7 col-xxl-6')


  # Country_box[0].find_all('h1')[0].text
  if len(Country_box[0].find_all('td'))==4:
    data=[]
    data.append(Country_box[0].find_all('h1')[0].text)

    for j in Country_box[0].find_all('td'):
      # print(j.text)

      data.append(j.text)
    print(counter,data[0])
    counter=counter+1
    Name.append(data[0])
    Population.append(data[1])
    CHE.append(data[2])
    WHO_REG.append(data[3])
    WBIL.append(data[4])

# data
```

```python
import pandas as pd

pd.DataFrame(
    {
        'Country':Name,
        'Population':Population,
        "Current Health Expenditure":CHE,
        'WHO REGION':WHO_REG,
        'World Bank income level':WBIL
    }
).to_csv('Data.csv')
```
